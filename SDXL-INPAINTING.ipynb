```python
from diffusers import AutoPipelineForInpainting
from diffusers.utils import load_image
import torch

pipe = AutoPipelineForInpainting.from_pretrained("diffusers/stable-diffusion-xl-1.0-inpainting-0.1", torch_dtype=torch.float16, variant="fp16").to("cuda")

img_url = "/content/1.JPG"
mask_url = "/content/mask_1.jpg"

image = load_image(img_url).resize((1024, 1024))
mask_image = load_image(mask_url).resize((1024, 1024))




prompt = "The aqueduct in the image is a long, stone-built structure that stretches across a dry, arid landscape. Constructed from large, irregularly shaped stones, the aqueduct's walls appear weathered and aged, with visible signs of erosion and decay. The stones are stacked without mortar, showcasing a Roman-era construction technique. The path of the aqueduct is a narrow, dirt-covered trail, flanked by the stone walls on either side. The trail is uneven, with small plants and patches of grass growing sporadically along its surface. The surrounding landscape is dry and sparse, with a few scattered trees and shrubs visible in the distance. The perspective of the image suggests that the viewer is standing at a higher elevation, looking down the length of the aqueduct as it recedes into the distance. The scale of the aqueduct is difficult to determine precisely, but it appears to be several meters wide and extends for several hundred meters. The texture of the stones is rough and uneven, with some areas showing significant wear and damage. The aqueduct's walls are slightly curved, following the natural contours of the landscape. Overall, the aqueduct is a striking example of ancient engineering and architecture, now partially reclaimed by the natural environment."
generator = torch.Generator(device="cuda").manual_seed(0)

image = pipe(
  prompt=prompt,
  image=image,
  mask_image=mask_image,
  guidance_scale=8.0,
  num_inference_steps=20,  # steps between 15 and 30 work well for us
  strength=0.99,  # make sure to use `strength` below 1.0
  generator=generator,
).images[0]
```
