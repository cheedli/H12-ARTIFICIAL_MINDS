```python
# Step 6: Create inference script with absolute paths
%%writefile run_inference.py
import os
os.environ['SPCONV_ALGO'] = 'native'
os.environ['ATTN_BACKEND'] = 'flash-attn'

import sys
import imageio
from PIL import Image
from trellis.pipelines import TrellisImageTo3DPipeline
from trellis.utils import render_utils, postprocessing_utils

def main(image_path):
    # Initialize pipeline
    pipeline = TrellisImageTo3DPipeline.from_pretrained("JeffreyXiang/TRELLIS-image-large")
    pipeline.cuda()

    # Load image
    image = Image.open(image_path)

    # Run inference with reduced parameters for Colab
    outputs = pipeline.run(
        image,
        seed=42,
        sparse_structure_sampler_params={
            "steps": 8,  # Reduced from 12
            "cfg_strength": 5.0  # Reduced from 7.5
        },
        slat_sampler_params={
            "steps": 8,  # Reduced from 12
            "cfg_strength": 2.0  # Reduced from 3
        }
    )

    # Save results with absolute paths
    output_dir = "/content/TRELLIS/outputs"
    os.makedirs(output_dir, exist_ok=True)

    for asset_type in ['gaussian', 'radiance_field', 'mesh']:
        video = render_utils.render_video(outputs[asset_type][0])['color']
        imageio.mimsave(f"{output_dir}/output_{asset_type}.mp4", video, fps=30)

    # Export 3D assets
    glb = postprocessing_utils.to_glb(outputs['gaussian'][0], outputs['mesh'][0])
    glb.export(f"{output_dir}/output.glb")

    outputs['gaussian'][0].save_ply(f"{output_dir}/output.ply")

if __name__ == "__main__":
    main(sys.argv[1])
```
